```python
from PIL import Image

img = Image.open(fp='tiger.jpeg')
img
# img.show(title='这是一张老虎图片')
```

```python
# 获取图像信息
print("格式:", img.format)  # JPEG/PNG
print("尺寸:", img.size)  # (宽, 高)
print("模式:", img.mode)  # RGB/L/CMYK
```

```python
gray_img = img.convert(mode='L')
gray_img
```

```python
img.thumbnail(size=(200, 200))
img
```

```python
cropped_img = img.crop(box=(0, 0, 100, 100))
cropped_img
```

```python


from PIL.Image import Transpose

img.transpose(method=Transpose.FLIP_TOP_BOTTOM)
```

```python
rotated_img = img.rotate(angle=45, expand=True)
rotated_img
```

```python
from PIL import ImageFilter

# 模糊滤镜
blurred_image = img.filter(filter=ImageFilter.BLUR)
blurred_image
```

```python
# 边缘增强
edge_enhanced = img.filter(filter=ImageFilter.EDGE_ENHANCE)
edge_enhanced
```

```python
from PIL import ImageEnhance

# 增强亮度
enhancer = ImageEnhance.Brightness(image=img)
bright_image = enhancer.enhance(factor=1.5)  # 1.5倍亮度
bright_image
```

```python
# 增强对比度
enhancer = ImageEnhance.Contrast(image=img)
contrast_image = enhancer.enhance(factor=2.0)  # 2倍对比度
contrast_image
```

```python
from PIL import ImageDraw

# 创建绘图对象
draw = ImageDraw.Draw(im=img)

# 绘制矩形
draw.rectangle(xy=[(100, 100), (300, 300)], outline="red")

# 绘制直线
draw.line(xy=[(0, 0), (500, 500)], fill="blue")

# 绘制圆形
draw.ellipse(xy=[(100, 100), (200, 200)], outline="green")

draw.circle(xy=(100, 100), radius=20, outline="red")

img
```

```python
from PIL import ImageFont

# 创建绘图对象
draw = ImageDraw.Draw(img)

# 加载字体 (需要系统中存在该字体)
font_size = 10  # 设置字体大小
try:
    font = ImageFont.truetype("arial.ttf", font_size)
except:
    font = ImageFont.load_default(font_size)
draw.text((10, 10), "Hello World", fill="red", font=font)

img
```

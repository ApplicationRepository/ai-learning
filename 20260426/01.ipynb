```python
new_df = df.dropna(axis=1)
new_df.select_dtypes(include=['int', 'float'])
new_df['SalePrice'].mean()
new_df['SalePrice'].median()
new_df.describe()
new_df.describe(include=['str'])
```

```python
new_df['PricePerSqFt'] = new_df['SalePrice'] / new_df['GrLivArea']

new_df[
    (new_df['PricePerSqFt'] > new_df['PricePerSqFt'].median())
    &
    (new_df['SalePrice'] < new_df['SalePrice'].median())
    ]
```

```python
new_df.groupby(by=['LotShape', 'LotConfig'])['SalePrice'].mean()
```

```python
new_df.groupby(by=['LotShape', 'LotConfig'])['SalePrice'].agg(func=['mean', 'median', 'std'])
```

```python
new_df.groupby(by=['LotShape', 'LotConfig']).agg(func={
    'Id': 'count',
    'SalePrice': ['mean', 'median', 'std'],
    'PricePerSqFt': ['mean', 'median', 'std']
}).to_excel('data.xlsx')
```

```python
new_df.sort_values(by=['LotShape', 'SalePrice'])
new_df.sort_values(by=['LotShape', 'SalePrice'], ascending=[True, False])
```

```python
from pptx import Presentation
from pptx.util import Inches


def create_ppt():
    # 1. 初始化演示文稿对象
    prs = Presentation()

    # --- 第一页：标题幻灯片 ---
    # slide_layouts[0] 通常是“标题”布局
    slide_layout = prs.slide_layouts[0]
    slide = prs.slides.add_slide(slide_layout)

    title = slide.shapes.title
    subtitle = slide.placeholders[1]

    title.text = "Python 自动化生成 PPT"
    subtitle.text = "使用 python-pptx 库的示例教程"

    # --- 第二页：项目符号列表 ---
    # slide_layouts[1] 通常是“标题和内容”布局
    bullet_slide_layout = prs.slide_layouts[1]
    slide = prs.slides.add_slide(bullet_slide_layout)

    shapes = slide.shapes
    title_shape = shapes.title
    title_shape.text = "功能特点"

    body_shape = shapes.placeholders[1]
    tf = body_shape.text_frame
    tf.text = "该库支持以下功能："

    # 添加一级项目符号
    p = tf.add_paragraph()
    p.text = "插入文本、图片和形状"
    p.level = 1

    # 添加二级项目符号
    p = tf.add_paragraph()
    p.text = "支持表格和图表自动化"
    p.level = 1

    # --- 第三页：插入图片 ---
    img_slide_layout = prs.slide_layouts[5]  # 只有标题的布局
    slide = prs.slides.add_slide(img_slide_layout)
    slide.shapes.title.text = "插入图片示例"
    # 设置图片位置和大小 (Inches 表示英寸)
    img_path = 'tiger.jpeg'  # 请确保路径下有这张图片，或替换为实际路径
    try:
        left = Inches(2)
        top = Inches(2)
        height = Inches(3)
        slide.shapes.add_picture(img_path, left, top, height=height)
    except Exception as e:
        print(f"图片插入失败（检查路径）: {e}")

    # --- 第四页：插入表格 ---
    table_slide_layout = prs.slide_layouts[5]
    slide = prs.slides.add_slide(table_slide_layout)
    slide.shapes.title.text = "数据展示表格"

    rows, cols = 3, 2
    left = Inches(2)
    top = Inches(2)
    width = Inches(6)
    height = Inches(0.8)

    table = slide.shapes.add_table(rows, cols, left, top, width, height).table

    # 设置列宽
    table.columns[0].width = Inches(2)
    table.columns[1].width = Inches(4)

    # 填充表头
    table.cell(0, 0).text = '项目'
    table.cell(0, 1).text = '数值'

    # 填充内容
    table.cell(1, 0).text = '自动化程度'
    table.cell(1, 1).text = '100%'
    table.cell(2, 0).text = '学习难度'
    table.cell(2, 1).text = '低'

    # 5. 保存文件
    prs.save('my_automation_ppt.pptx')
    print("PPT 已生成成功！")


if __name__ == "__main__":
    create_ppt()
```
